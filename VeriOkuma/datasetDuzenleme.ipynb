```python
import pandas as pd
```

```python
# reading csv files
data =  pd.read_csv('Covid Data.csv', sep=",")
df = data.copy() #her ihtimale karşı yedekleyelim
df.head()
```

```python
df.columns = ["TedaviDurumu","TibbiBirim","Cinsiyet","HastaTuru","SaglikDurumu","Entube","Zaturre","Yas","Hamile","Diyabet",
             "KOAH","Astim","BagisiklikSistemi","HiperTansiyon","DigerHastalik","Kardiyovaskuler","Obezite","KronikBobrek",
             "Sigara","CovidSiniflandirma","ICU"] 
#değişken isimlerini değiştiriyoruz
```

```python
df.info() 
#kaba bir şekide veri setinin tüm yapısal bilgilerini verir.
#bir pandas dataframe şeklindeymiş
#53940 adet gözlem değerimiz varmış
# 10 değişkenimiz varmış, değişkenlerin tipleri belirtilmiş
#Ramda 3MB alan tutmuş
```

```python
df
```

```python
# 'cinsiyet' sütunundaki değerleri değiştir
df['Cinsiyet'] = df['Cinsiyet'].map({1: 'Kadın', 2: 'Erkek'})
```

```python
df
```

```python
# Cinsiyeti 1 olanları kadın yap, değilse erkek yap
df.loc[df['SaglikDurumu'] != '9999-99-99', 'SaglikDurumu'] = 'Ölü'
df.loc[df['SaglikDurumu'] == '9999-99-99', 'SaglikDurumu'] = 'Sağ'
```

```python
df
```

```python
df['Entube'] = df['Entube'].astype(str)  #Null yapmak için ilk önce değişken türünü null yapıyorum
df['TedaviDurumu'] = df['TedaviDurumu'].astype(str) 
df['TibbiBirim'] = df['TibbiBirim'].astype(str) 
df['Cinsiyet'] = df['Cinsiyet'].astype(str) 
df['HastaTuru'] = df['HastaTuru'].astype(str) 
df['SaglikDurumu'] = df['SaglikDurumu'].astype(str)
df['Zaturre'] = df['Zaturre'].astype(str) 
df['Yas'] = df['Yas'].astype(str) 
df['Hamile'] = df['Hamile'].astype(str) 

df['KOAH'] = df['KOAH'].astype(str) 
df['Astim'] = df['Astim'].astype(str) 
df['BagisiklikSistemi'] = df['BagisiklikSistemi'].astype(str) 
df['HiperTansiyon'] = df['HiperTansiyon'].astype(str)
df['DigerHastalik'] = df['DigerHastalik'].astype(str) 
df['Kardiyovaskuler'] = df['Kardiyovaskuler'].astype(str) 
df['Obezite'] = df['Obezite'].astype(str) 

df['KronikBobrek'] = df['KronikBobrek'].astype(str) 
df['Sigara'] = df['Sigara'].astype(str) 
df['CovidSiniflandirma'] = df['CovidSiniflandirma'].astype(str) 
df['ICU'] = df['ICU'].astype(str)
```

```python
df
```

```python
df.loc[df['TedaviDurumu'] == '99', 'TedaviDurumu'] = None 
df.loc[df['TedaviDurumu'] == '97', 'TedaviDurumu'] = None 
df.loc[df['TibbiBirim'] == '99', 'TibbiBirim'] = None 
df.loc[df['TibbiBirim'] == '97', 'TibbiBirim'] = None 
df.loc[df['Cinsiyet'] == '99', 'Cinsiyet'] = None 
df.loc[df['Cinsiyet'] == '97', 'Cinsiyet'] = None 
df.loc[df['HastaTuru'] == '99', 'HastaTuru'] = None 
df.loc[df['HastaTuru'] == '97', 'HastaTuru'] = None 
df.loc[df['SaglikDurumu'] == '99', 'SaglikDurumu'] = None 
df.loc[df['SaglikDurumu'] == '97', 'SaglikDurumu'] = None 
df.loc[df['Entube'] == '99', 'Entube'] = None 
df.loc[df['Entube'] == '97', 'Entube'] = None 
df.loc[df['Zaturre'] == '99', 'Zaturre'] = None 
df.loc[df['Zaturre'] == '97', 'Zaturre'] = None 
df.loc[df['Yas'] == '99', 'Yas'] = None 
df.loc[df['Yas'] == '97', 'Yas'] = None 
df.loc[df['Hamile'] == '99', 'Hamile'] = None 
df.loc[df['Hamile'] == '97', 'Hamile'] = None 
df.loc[df['Diyabet'] == '99', 'Diyabet'] = None 
df.loc[df['Diyabet'] == '97', 'Diyabet'] = None 
df.loc[df['KOAH'] == '99', 'KOAH'] = None 
df.loc[df['KOAH'] == '97', 'KOAH'] = None 
df.loc[df['Astim'] == '99', 'Astim'] = None 
df.loc[df['Astim'] == '97', 'Astim'] = None 
df.loc[df['BagisiklikSistemi'] == '99', 'BagisiklikSistemi'] = None 
df.loc[df['BagisiklikSistemi'] == '97', 'BagisiklikSistemi'] = None 
df.loc[df['HiperTansiyon'] == '99', 'HiperTansiyon'] = None 
df.loc[df['HiperTansiyon'] == '97', 'HiperTansiyon'] = None 
df.loc[df['DigerHastalik'] == '99', 'DigerHastalik'] = None 
df.loc[df['DigerHastalik'] == '97', 'DigerHastalik'] = None 
df.loc[df['Kardiyovaskuler'] == '99', 'Kardiyovaskuler'] = None 
df.loc[df['Kardiyovaskuler'] == '97', 'Kardiyovaskuler'] = None 
df.loc[df['Obezite'] == '99', 'Obezite'] = None 
df.loc[df['Obezite'] == '97', 'Obezite'] = None 
df.loc[df['KronikBobrek'] == '99', 'KronikBobrek'] = None 
df.loc[df['KronikBobrek'] == '97', 'KronikBobrek'] = None 
df.loc[df['Sigara'] == '99', 'Sigara'] = None 
df.loc[df['Sigara'] == '97', 'Sigara'] = None 
df.loc[df['CovidSiniflandirma'] == '99', 'CovidSiniflandirma'] = None 
df.loc[df['CovidSiniflandirma'] == '97', 'CovidSiniflandirma'] = None
df.loc[df['ICU'] == '99', 'ICU'] = None 
df.loc[df['ICU'] == '97', 'ICU'] = None 
```
